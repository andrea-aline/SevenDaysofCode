### Step 4 - Remove unnecessary columns

```python
#Remove unnecessary columns
data.drop(columns = ['DOCUMENTO','COD_DOCUMENTO','CNPJ_CPF'],axis=1,inplace=True)
```

```python
data.head(2)
```

### Step 5 - make sure all values in 'VALOR_REEMBOLSADO' are numeric

```python
#Make sure all values in 'VALOR_REEMBOLSADO' are numeric
valor_series_num = pd.Series(data['VALOR_REEMBOLSADO'])
valor_series_num.str.isnumeric()
```

```python
#See one example of non-numeric
data['VALOR_REEMBOLSADO'].iloc[16822]
```

It might have something to do with the decimal separator

```python
#When I try to calculate a simple sum with 'VALOR_REEMBOLSADO' in line 16822. it returns an error
data['VALOR_REEMBOLSADO'].iloc[16822] + 1
```

```python
#However, when I try to concatenate a string, it works
data['VALOR_REEMBOLSADO'].iloc[16822] + 'A'
```

```python
#Replace "," to "." as decimal separator using list comprehension
data['VALOR_REEMBOLSADO'] = [x.replace(',','.') for x in data['VALOR_REEMBOLSADO']]
```

```python
#Make sure 'VALOR_REEMBOLSADO' is a float
data['VALOR_REEMBOLSADO'] = data['VALOR_REEMBOLSADO'].astype(float)
```

```python
#Check if the change was applied
data['VALOR_REEMBOLSADO'].iloc[16822]
```

```python
#Now, the calculation works
data['VALOR_REEMBOLSADO'].iloc[16822] + 1
```
